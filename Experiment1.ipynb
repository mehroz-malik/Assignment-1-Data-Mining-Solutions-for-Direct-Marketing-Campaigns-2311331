# Experiment 1: Baseline Decision Tree



```r
# Load the dataset
bank_data <- read.csv("bank-additional.csv" , sep = ";")
```

```r
# Preprocess the data
bank_data$y <- as.factor(bank_data$y)
set.seed(123)
library(caret)
split <- createDataPartition(bank_data$y, p = 0.8, list = FALSE)
train_data <- bank_data[split, ]
test_data <- bank_data[-split, ]
```

```r
# Building the Decision Tree model
library(rpart)
tree_model <- rpart(y ~ ., data = train_data, method = "class")
```

```r
# Visualize the tree
library(rpart.plot)
rpart.plot(tree_model)
```

```r
# Evaluate the model
predictions <- predict(tree_model, test_data, type = "class")
confusionMatrix(predictions, test_data$y)
```
